```python
x=[0.75, 0.65, 0.80]
w=[0.4, 0.3, 0.5]
b=-0.2
Y=1
n=0.01
z=0
for i in range(3):
    z+=x[i]*w[i]
z=z+b
print("Net input(z):", z)
if z>= 0:
    y_pred = 1
else:
    y_pred = 0
print("Predicted output:", y_pred)
error = Y-y_pred
print("Error:", error)
for i in range(3):
    w[i] = w[i]+n*error*x[i]
b=b+n*error
print("Updated weights:", w)
print("Updated bias:", b)
```
